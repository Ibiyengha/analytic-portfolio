```python
import matplotlib.pyplot as plt #for visualization
import seaborn as sb #use to visualize ranndom distribution
```

#LIMIT_BAL

```python
fig, axes = plt.subplots(1,2, figsize=(10,4))
sb.histplot(credit_card['LIMIT_BAL'], bins=10,ax=axes[0])
plt.boxplot(credit_card['LIMIT_BAL'],patch_artist = True)

axes[0].set_title('Distribution of the Amount of Given Credit of Customers')
axes[1].set_title('Range of the Amount of Given Credit   of Customers')
plt.tight_layout()
# plt.savefig("hist of LIMIT_BALofCustomers.png")
plt.show()
```

Looking at the boxplot, Outliers were detected in LIMIT_BAL attribute, so i will be removing these outliers by calculating the limits on the variable and taking outliers as observations that fall outside of the range.

```python
 # calculating Q1 for LIMIT_BAL attribute
# 0.25 means 25% on left and 75% on right
Q1_LIMIT_BAL = credit_card["LIMIT_BAL"].quantile(0.25)
print("Q1_LIMIT_BAL  :", int(Q1_LIMIT_BAL ),"NT dollars")
# calculating Q3 for LIMIT_BAL attribute
Q3_LIMIT_BAL   = credit_card["LIMIT_BAL"].quantile(0.75)
print("Q3_LIMIT_BAL :", int(Q3_LIMIT_BAL ),"NT dollars")
# calculate interquartile (IQR) for LIMIT_BAL attribute
IQR_LIMIT_BAL  = Q3_LIMIT_BAL   - Q1_LIMIT_BAL
print("IQR_LIMIT_BAL", int(IQR_LIMIT_BAL ),"NT dollars")
#calculating lower bound for LIMIT_BAL atrribute
lowerBound_LIMIT_BAL  = Q1_LIMIT_BAL  - (1.5 * IQR_LIMIT_BAL)
lowerBound_LIMIT_BAL
#calculating upper bound for LIMIT_BAL atrribute
upperBound_LIMIT_BAL   = Q3_LIMIT_BAL + (1.5 * IQR_LIMIT_BAL)
upperBound_LIMIT_BAL
```

```python
#removing the outlier
#The goal is to eliminate outliers from  'LIMIT_BAL'  where values exceed the specified upper bound value.
index = credit_card[(credit_card['LIMIT_BAL']>upperBound_LIMIT_BAL)].index
credit_card = credit_card.drop(index)
```

```python
#visualizing distribution of LIMIT_BAL attribute
sb.boxplot(x = 'LIMIT_BAL', data = credit_card)
plt.tight_layout()
#plt.savefig("hist of LIMIT_BAL.png")
plt.show()
```
